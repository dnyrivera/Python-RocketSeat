```python
mylist = [1,2,3,4,5,"rocketseat",True,False]
```

```python
print(f"Minha lista de exemplo: {mylist}")
```

```python
print("my_list[0]:", mylist[0])
print("my_list[5]:", mylist[5])
print("my_list[1:7]:", mylist[1:7])
print("my_list[:5]:",mylist[:6])
print("my_list[2:]:",mylist[2:])
print("my_list[-1]:",mylist[-1])
```

```python
mylist[0] = "python"
print(mylist)
```

```python
# Methods

# Append 
mylist.append(6)
print("After Append:", mylist)
```

```python
# Index
print(mylist.index(6))
```

```python
# Insert
mylist.insert(0,"first")
print(mylist)
```

```python
# Pop
# mylist.pop()
removed_element = mylist.pop(3)
print(f"Removed Element: {removed_element}")
print(f"After Pop(3); {mylist}")

# mylist.remove(True)
# print(mylist)
```
