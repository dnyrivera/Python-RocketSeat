```python
print("Example")
try:
    number = int(input("Type a number: "))
    result = 10 / number
except ValueError as e:
    print(f"Erro de value erro {e}")
    raise ValueError("Variable Type Incompatible")
except Exception as e:
    print(f"Erro: {e}")
else:
    print(f"{result}")
finally:
    print("Final Processing")
```
