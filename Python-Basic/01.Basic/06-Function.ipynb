```python
def hello(name:str) -> None:
    print(f"Hi, {name}")

print ("Calling the function Hello")
hello("Alice")
hello("Bob")
```

```python
# Returning Values 
def square (number:int) -> int:
    result = number ** 2 
    return result

print("Square Function")
square_result = square(40)
print(f"Square Total: {square_result}")
```

```python
# Multiples Parameters 

def sum (number_one: float, number_two: float) -> float:
    result = number_one + number_two
    return result

print("Call the Sum Function")
result_sum = sum(13.5, 50)
print(f"Sum Total: {result_sum}")
```
