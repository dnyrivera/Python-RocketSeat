```python
 # Create Tuple

my_tuple = (1, 2, 2, 3, 3, 4, 4, 5, 5)
print(my_tuple)
```

```python
print(f"My Tuple[0]:{my_tuple[0]}")
print(f"My Tuple[1]:{my_tuple[1]}")
print(f"My Tuple[-1]:{my_tuple[-1]}")
```

```python
# Count 
count = my_tuple.count(5)
print(f"How many time the element 5 show in my list: {count}")
```
