```python
# For
print('Printing each item for one List')
list = [1, 2, 3, 4, 5]
for elements in list:
    print(elements, end=',')
```

```python
print('Printing each item for one Tuple')
tuple = ('a', 'b', 'c', 'd', 'e')
for elements in tuple:
    print(elements,  end='-')
```

```python
print('Printing each item for one Dictionary\n')
people = {
    "name": "Donny",
    "last name": "Rivera",
    "age": 42
}

print('Printing the Dictionary Keys()')
for keys in people.keys():
    print(keys, end=' | ')

print('\n \nPrinting the Dictionary Values()')
for values in people.values():
    print(values, end=' | ')

print('\n \nPrinting the Dictionary Keys-Values()')
for keys, values in people.items():
    print(f"{keys}: {values}", end=' | ')
```

```python
# Range() - between numbers
print('\nFor using range()')
for number in range(5):
    print('Num: ', number)

# len() - size of the list
print('\nFor using range() and len()')
list = [1, 2, 3, 4, 5, 6 , 7 , 8]
for index in range(0, len(list)):
    print(f'Pos: {index} | Num: {list[index]}')
```

```python
print('\nFor using enumerate(list)')
list_enum = ['a', 'b', 'c']
for index , value in enumerate(list_enum):
    print(f"{index}:{value}")
    if index == 2: 
        list_enum[index] = 'd'

print(list_enum)
```
