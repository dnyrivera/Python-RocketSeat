```python
# decisions (if - elif - else)

age = 19
print("IF - Example")
if age >= 18:
    print('You are an adult')

if age == 19:
    print('You are 19 years old')

if age < 18:
    print("You are a teenage or a child")

if age != 10:
    print("You aren't 10 years.")
```

```python
# decisions (if - elif - else)

age = 11
print("ELIF - Example")
if age >= 18:
    print('You are an adult')
elif age >= 12:
    print("You are a teenage")
else:
    print("You are a child")
```

```python
# decisions (if - elif - else)

age = 18
print("ELSE - Example")
if age >= 18:
    print('You are an adult')
else:
    print("You are a teenage or a child")
```

```python
# conditional in one line 

message = 'You can have you drive license' if age >= 18 else "You can't have a drive license"
print(message)
```
