```python
# create dictionary
people = {"name": "Donny", "age": 42, "weight": 105, "city": "Toronto", }
```

```python
# print dictionary
print(people)
```

```python
# print key values
print(
    f"Name: {people['name']}\n"
    f"Age: {people['age']}\n"
    f"Weight: {people['weight']}\n"
    f"City: {people['city']}"
)
```

```python
# adding a new key
people['last-name'] = 'Rivera'
print(f"Last Name: {people['last-name']}", people, sep="\n")
```

```python
# update keys
people['age'] = 43
print(f"Age Updated: {people}")
```

```python
# delete key / value
deleted_key = people.pop("last-name")
print(f"Key last-name deleted. Value was: {deleted_key}")
print(f"Updated dictionary: {people}")
```

```python
# Methods: keys(), values() and items()
keys_dict = list(people.keys())  # casting to list the keys like a list
print(f"Dict Keys: {keys_dict}")
print(f"First Key: {keys_dict[0]}")
```

```python
values_dict = list(people.values())
print(f"Dict Values: {values_dict}")
print(f"First Valeu: {values_dict[0]}")
```

```python
items_dict = list(people.items())
print(f"Dict Items: {items_dict}")
print(f"First Key-Value: {items_dict[0][0]} - {items_dict[0][1]}")
```
